```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery')

# make data
x = [.1, .3, .5]
y = [.1, .4, .7]
y2 = [.0, .6, .9]

# plot
fig, ax = plt.subplots()

ax.plot(x, y, linewidth=2.0)
ax.plot(x, y2, linewidth=2.0)


ax.set(xlim=(0, 1), xticks=np.arange(0, 1.1, .2),
       ylim=(0, 1), yticks=np.arange(.2, 1.1, .2))

ax.set_xlabel('Recall')
ax.set_ylabel('Precession')
ax.set_title('11 point')
ax.legend(['vsm', 'rm25', 'lm'], title="model")

plt.show()
```
